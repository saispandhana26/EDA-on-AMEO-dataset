# Dataset Description
1. The Aspiring Mind Employment Outcome 2015 (AMEO) dataset, released by Aspiring Minds, focuses on employment outcomes for engineering graduates. It includes dependent variables such as Salary, Job Titles, and Job Locations, along with standardized scores in cognitive skills, technical skills, and personality skills. With around 40 independent variables and 4000 data points, these variables encompass both continuous and categorical data. The dataset also includes demographic features and unique identifiers for each candidate.

# Objective
1. The goal of this Exploratory Data Analysis (EDA) is to extensively investigate the provided dataset, with a particular emphasis on understanding the link between various variables and the target variable, Salary.

2. The key aims of this analysis include:

3. Providing a detailed explanation of the dataset's features.
4. Find any observable patterns or trends in the data.
5. Investigating the relationships between the independent factors and the target variable (salary).
6. Identify any outliers or abnormalities in the dataset.
7. Offering practical insights and recommendations based on the analysis.

# converting the columns with dates to datetime

```python
# converting date of leaving having value as "present" to present date i.e 31-12-2015 as the survey was conducted back in 201
df1['DOL'].replace('present','31/12/15', inplace = True)
df1['DOL'] = pd.to_datetime(df1['DOL'])
df1.head()
```

### one more concern is that the date of leaving shouldn't be earlier than date of joining if there are such values they might be typos so we will drop such rows

```python
df1 = df1.drop(columns = ['MechanicalEngg', 'ElectricalEngg', 'TelecomEngg', 'CivilEngg'])
df1.head()
```

```python
df1['10board'] = df1['10board'].replace({'0':np.nan})
df1['12board'] = df1['12board'].replace({'0':np.nan})
df1['GraduationYear'] = df1['GraduationYear'].replace({0:np.nan})
df1['JobCity'] = df1['JobCity'].replace({'-1':np.nan})
df1['Domain'] = df1['Domain'].replace({-1:np.nan})
df1['ElectronicsAndSemicon'] = df1['ElectronicsAndSemicon'].replace({-1:np.nan})
df1['ComputerScience'] = df1['ComputerScience'].replace({-1:np.nan})
df1['ComputerProgramming'] = df1['ComputerProgramming'].replace({-1:np.nan})
```

## converting all the textual  columns into lower case

```python
def lower_data(data):
    df1[data] = df1[data].str.lower().str.strip()
```

```python
for col in t_columns:
    lower_data(col)
```

## since there are large num of categories in each column lets only take top 20 values and keep the other values as "others"

```python
def prominent_categories(df1, data):
    for Designation in df1[data].unique():
        min_count = df1[data].value_counts()[:15].min()
        if df1[df1[data] == Designation][data].value_counts()[0] < min_count:
            df1.loc[df1[data] == Designation, data] = 'other'
```

```python
for cols in t_columns:
    prominent_categories(df1, cols)
```
